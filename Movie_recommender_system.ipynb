```python
#we will import all the neccessary libraries and packages
import numpy as np
import pandas as pd
```

```python
#loading the data
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
#merging two dataframes to make our task e=comparatively easier.
movies = movies.merge(credits , on = 'title')
```

```python
movies.head(1)
```

```python
#important features that we are going to consider
#movie id
#title
#genres
#keywords
#overview
#cast
#crew
movies = movies[['movie_id','title','genres','keywords','overview','cast','crew']]
        
```

```python
movies.head(1)
```

```python
#checking if there are any missing values present in the data
movies.isnull().sum()
```

```python
#excluding/droping the missing data
movies.dropna(inplace= True)
```

```python
#checking if data contains any duplicates
movies.duplicated().sum()
```

```python
movies.head()
```

```python
#accessing data based on integers position
movies.iloc[0].genres
```

```python
# these infos are in string format, but we have convert this in int format like 
# ['Action' , 'Adventure' , 'Fantasy' , 'Science Fiction']
#hum genre ko ek list banana chahte hai but wo string me present hai toh pehle use int me covert krna pdega tabhi ye code chlega toh uske lie 
# ast literal eval ka use karte hai.
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#for converting cast into list format
def convert1(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert1)
```

```python
movies.head()
```

```python
#for converting crew into list format
def director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L 
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
# to break paragraph into list
movies['overview'] = movies['overview'].apply(lambda x:x.split()) 
```

```python
movies.head()
```

```python
movies['overview']
```

```python
#transformation to remove spaces btw two words so that recommender system could work perfectly

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
df = movies[['movie_id','title','tags']]
```

```python
df
```

```python
# just a conversion again from list to string
df['tags'] = df['tags'].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df['tags'][0]
```

```python
#its simply just recommended to covert it in lowercase
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
import nltk
```

```python
#predefined rules to convert words into their root forms
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)   
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
df['tags'][0]
```

```python
#transform text data into numerical data that can be used by machine learning models.

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
## we will find cosine similarity where cosine dist is inversely proportional to similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse = True ,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = df[df['title']== movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True ,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(df.iloc[i[0]].title)
    
    
```

```python
recommend('Falcon Rising')
```
